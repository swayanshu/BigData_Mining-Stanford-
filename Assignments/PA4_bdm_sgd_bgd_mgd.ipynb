```python
import time
import numpy as np
import os
import copy
```

```python
# opening file
def file_length(filename):
    with open(filename) as f:
        for i, l in enumerate(f):
            pass
    return i + 1
```

```python
def feature_length_calculation(filename):
    with open(filename) as fn:
        length = len(fn.readline().split(","))
    return length
```

```python
C = 100
feature = "features.txt"
target = "target.txt"
length = file_length(feature)
f_len = feature_length_calculation(feature)
```

```python
length
```

```python
# gradient descent
feature_file = open(feature, "r")
target_file = open(target, "r")
```

```python
# the last element in each line of entries is the target
datasets = []
for file in feature_file.readlines():
    list1 = file.strip().split(",")
    datasets.append(list(map(int, list1)))

#enumerating target file
for (row, column) in enumerate(target_file):
    target_data = column.strip()
    datasets[row].append(int(target_data))
```

# Batch gradient descent

```python
k_bgd = 0
v = np.zeros(f_len) 
v_grad = np.zeros(f_len)
learning_rate = 0.0000003
epsilon = 0.25
m = 0.0
m_grad = 0.0
fk_bgd = [C * length]
```

```python
start = time.process_time()
while True:
    sum_two = np.zeros(f_len)
    for data in datasets:
        a = np.array(data[:-1])
        b = np.array(data[-1])
        av = sum(a * v)
        equation = b * (av + m)
        if equation >= 1:
            pass
        else:
            sum_two -= b * a
    v_new = v - learning_rate * (v + C * sum_two)
# New batch 
    new_b = 0.0
    for data in datasets:
        a = data[:-1]
        b = data[-1]
        ab = sum(a * v)
        condition = b * (ab + m)
        if condition >= 1:
            pass
        else:
            new_b -= b
    new_m1 = m - learning_rate * C * new_b
 # update w and b
    v = copy.deepcopy(v_new)
    m = copy.deepcopy(new_m1)
 # calculate error
    eq = 0.5 * sum(v * v)
    sec_i = 0.0
    for data in datasets:
        a = data[:-1]
        b = data[-1]
        ab = sum(a * v)
        condition = b * (ab + m)
        if condition >= 1:
            pass
        else:
            sec_i += 1 - condition
    eq += C * sec_i
    fk_bgd.append(eq)
# delta calculation
    if 0 == k_bgd:
        pass
    else:
        delta_updated = abs(fk_bgd[k_bgd - 1] - fk_bgd[k_bgd]) / fk_bgd[k_bgd - 1] * 100
        if delta_updated <= epsilon:
            break
    k_bgd += 1
#calculate how much time taken
finish = time.process_time()
bgd_e = finish - start
print (bgd_e)
```

# SGD

```python
np.random.shuffle(datasets)
```

```python
k_iteration = 0
i = 0
v = np.zeros(f_len) 
b = 0.0
f_sgd = [C * length]
d_file = [0]
learning_rate = 0.0001
epsilon = 0.001
```

```python
start_time = time.process_time()
while True:
    sum_two = np.zeros(f_len)
    a = np.array(datasets[i][:-1])
    b = np.array(datasets[i][-1])
    av = sum(a * v)
    equation = b * (av + m)
    if equation >=1:
        pass
    else:
        sum_two -= b * a
    v_new = v - learning_rate * (v + C * sum_two)

    # calculate new b using old b, w
    new_b = 0.0
    a = datasets[i][:-1]
    b = datasets[i][-1]
    av = sum(a * v)
    equation = b * (av + m)
    if equation >= 1:
        pass
    else:
        new_b -= b
    new_m1 = m - learning_rate * C * new_b

    # update b and w
    v = copy.deepcopy(v_new)
    m = copy.deepcopy(new_m1)

    # calculate error
    eq = 0.5 * sum(v * v)
    sec_i = 0.0
    for data in datasets:
        a = data[:-1]
        b = data[-1]
        ab = sum(a * v)
        condition = b * (ab + m)
        if condition >= 1:
            pass
        else:
            sec_i += 1 - condition
    eq += C * sec_i
    f_sgd.append(eq)
  # calculate delta
    if 0 == k_iteration:
        pass
    elif 1 == k_iteration:
        delta = abs(f_sgd[k_iteration - 1] - f_sgd[k_iteration]) / f_sgd[k_iteration - 1] * 100
        d_file.append(delta)
    else:
        delta = abs(f_sgd[k_iteration - 1] - f_sgd[k_iteration]) / f_sgd[k_iteration - 1] * 100
        delta_curr = 0.5 * d_file[k_iteration - 1] + 0.5 * delta
        d_file.append(delta_curr)
        if delta_curr <= epsilon:
            break
    k_iteration += 1
    i = (i + 1) % length

end = time.process_time()
finished_time = end - start_time
print (finished_time)
```

# mgd

```python
np.random.shuffle(datasets)
```

```python
k_mgd = 0
n = 0
v = np.zeros(f_len) 
b = 0.0
fk_mgd = [C * length]
delta_list_mgd = [0]
learning_rate = 0.00001
epsilon = 0.01
split = 20
```

```python
start_t = time.process_time()

while True:
    mini_batch_gradient = copy.deepcopy(datasets[int(n * split) : min(length, int((n + 1) * split))])

    s_num = np.zeros(f_len)
    for data in mini_batch_gradient:
        a = np.array(data[:-1])
        b1 = np.array(data[-1])
        av = sum(a * v)
        equation_mgd = b1 * (av + b)
        if equation_mgd >= 1:
            pass
        else:
            s_num -= b1 * a
    v_new = v - learning_rate * (v + C * s_num)
    b_sum = 0.0
    for entry in mini_batch_gradient:
        a = entry[:-1]
        b1 = entry[-1]
        av = sum(a * v)
        updated = b1 * (av + b)
        if updated >= 1:
            pass
        else:
            b_sum -= b1
    b_new = b - learning_rate * C * b_sum

    # update w and b
    v = copy.deepcopy(v_new)
    m = copy.deepcopy(b_new)

    # calculate error
    f_temporary = 0.5 * sum(v * v)
    sec_i = 0.0
    for data in datasets:
        a = data[:-1]
        b1 = data[-1]
        av = sum(a * v)
        updated = b1 * (av + b)
        if updated >= 1:
            pass
        else:
            sec_i += 1 - updated
    f_temporary += C * sec_i
    fk_mgd.append(f_temporary)
  # calculate delta
    if 0 == k_mgd:
        pass
    elif 1 == k_mgd:
        delta = abs(fk_mgd[k_mgd - 1] - fk_mgd[k_mgd]) / fk_mgd[k_mgd - 1] * 100
        delta_list_mgd.append(delta)
    else:
        delta = abs(fk_mgd[k_mgd - 1] - fk_mgd[k_mgd]) / fk_mgd[k_mgd - 1] * 100
        delta_current = 0.5 * delta_list_mgd[k_mgd - 1] + 0.5 * delta
        delta_list_mgd.append(delta_current)
        if delta_current <= epsilon:
            break
    k_mgd += 1
    n = (n + 1) % ((length + split - 1) / split)

end_t = time.process_time()
total_time_mgd = end_t - start_t
print (total_time_mgd)
```
